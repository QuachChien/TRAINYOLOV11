```python
from ultralytics import YOLO
import os

# Sửa đường dẫn này cho đúng
dataset_path = "/content/drive/MyDrive/ft456_/ft465_L_0423.v1i.yolov11"   # ← SỬA TÊN FOLDER Ở ĐÂY

data_yaml = os.path.join(dataset_path, "data.yaml")

print("Đường dẫn data.yaml:", data_yaml)
print("Tồn tại?", os.path.exists(data_yaml))

# Kiểm tra cấu trúc
!ls "{dataset_path}"
```

```python
from ultralytics import YOLO

data_yaml = "/content/drive/MyDrive/ft456_/ft465_L_0423.v1i.yolov11/data.yaml"

model = YOLO("/content/yolo11n.pt")

results = model.train(
    data=data_yaml,
    epochs=100,
    imgsz=640,
    batch=64,           # Tăng lên 32 (thử trước)
    patience=50,
    cache=True,
    workers=8,          # Tăng worker
    name="yolo11_ft465_20260704_v2",
    exist_ok=True,
    amp=True
)

# Backup tự động
!cp -r /content/runs/detect/yolo11_ft465_20260704 /content/drive/MyDrive/yolo11_backup/
print("✅ Đã backup vào Drive!")
```
